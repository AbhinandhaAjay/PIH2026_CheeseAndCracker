```python
import os

# Define the new mapping (index start from 18 for dataset2 classes)
new_class_offset = 11

# Paths
label_dir = 'dataset4/train/labels'

# Update the labels
for file_name in os.listdir(label_dir):
    if file_name.endswith('.txt'):
        file_path = os.path.join(label_dir, file_name)
        with open(file_path, 'r') as file:
            lines = file.readlines()
        with open(file_path, 'w') as file:
            for line in lines:
                parts = line.strip().split()
                class_id = int(parts[0]) + new_class_offset
                new_line = ' '.join([str(class_id)] + parts[1:]) + '\n'
                file.write(new_line)

print("Dataset labels remapped!")
```

```python
unified_yaml = """
train: ../merged/train/images
val: ../merged/valid/images
test: ../merged/test/images

nc: 13
names: ['ModerateAccident', 'NoAccident', 'SevereAccident','injured', 'notinjured', 'Car', 'Jeep', 'Motorcycle', 'Tricycle', 'Truck', 'Van', 'fire', 'smoke']

"""

with open("/content/merged/unified_data.yaml", "w") as f:
    f.write(unified_yaml)
```

```python
from ultralytics import YOLO

# Load YOLOv8 model
model = YOLO("yolov8n.pt")  # You can replace "yolov8n.pt" with other versions like yolov8s.pt

# Train the model
model.train(data="/content/merged/unified_data.yaml", epochs=20, imgsz=640, batch=32)
```

```python
results = model.val()
```
